```python
%matplotlib inline # получать результаты визуализаций в юпитер-ноутбуке
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students_performance = pd.read_csv('StudentsPerformance.csv')
students_performance.columns = [x.replace(" ", "_") for x in students_performance.columns]
students_performance.head()
```

### seaborn
https://seaborn.pydata.org/introduction.html

https://seaborn.pydata.org/generated/seaborn.lmplot.html?highlight=seaborn%20lmplot#seaborn.lmplot

```python
sns.lmplot(x='math_score', y='reading_score', data=students_performance)
```

добавим группирующую переменную:

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance)
```

без регрессионой прямой:

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg = False)
```

кастомизация:

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg = False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
