```python
students_performace = pd.read_csv("C:/Users/TAGIR/Documents/StudentsPerformance.csv")
```

```python
mean_writing_score = students_performace["writing score"].mean()
```

```python
students_performace.loc[students_performace["writing score"] > mean_writing_score].head()
```

```python
students_performace.loc[students_performace["lunch"] == "free/reduced"].describe()
```

```python
students_performace.loc[students_performace["lunch"] == "standard"].describe()
```

```python
students_performace = students_performace\
                                    .rename(columns=
                                            {'parental level of education': 'parental_level_of_education',
                                            'test preparation course': 'test_preparation_course',
                                            'math score': 'math_score',
                                            'reading score': 'reading_score',
                                            'writing score': 'writing_score'})

#students_performance.columns=students_performance.columns.str.replace(' ','_')
```

```python
students_performace.query("gender == 'female' & writing_score > 78")
```

```python
score_columns = [i for i in list(students_performace) if "score" in i]
```

```python
students_performace[score_columns].head()
```

```python
students_performace.filter(like = "score")
```

```python
students_performace.groupby('lunch').mean()
```

```python
students_performace.groupby("lunch", as_index = False) \
    .aggregate({"math_score" : "mean", "reading_score" : "mean"}) \
    .rename(columns = {"math_score" : "mean_math_score", "reading_score" : "mean_reading score"})
```

```python
students_performace.sort_values(['gender', 'math_score'], ascending=[True, False])
```

```python
students_performace["math_score"].unique()
```

```python
lunch_group = students_performace.groupby(["lunch"]).math_score.unique()
```

```python
type(lunch_group)
```

```python
# топ 5 студентов ао матем.
students_performace.sort_values(["gender", "math_score"], ascending = False) \
    .groupby("gender").head(5)
```

```python
students_performace["total_score"] = students_performace.math_score + students_performace.reading_score \
    + students_performace.writing_score
```

```python
students_performace = students_performace.assign(total_score_log = np.log(students_performace.total_score))
```
